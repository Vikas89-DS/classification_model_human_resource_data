```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data

employee_df = pd.read_csv("Human_Resources.csv")
employee_df.head(5)
```

```python
#checking data for null values and data types

employee_df.info()
```

```python
# Converting 'Attrition' Column to Binary Values

employee_df["Attrition"] = employee_df["Attrition"].apply(lambda x:1 if x == "Yes" else 0)
```

```python
# Converting 'Over18' Column to Binary Values

employee_df["Over18"] = employee_df["Over18"].apply(lambda x:1 if x == "Y" else 0)
```

```python
# Converting 'OverTime' Column to Binary Values

employee_df["OverTime"] = employee_df["OverTime"].apply(lambda x:1 if x == "Yes" else 0)
```

```python
employee_df
```

```python
# Ploting Histograms for All Numerical Features

employee_df.hist(bins =30, figsize = (20,20), color = "b")
```

```python

drop_list = ["EmployeeCount", "Over18", "StandardHours", "EmployeeNumber"]
```

```python
#droping non required columns
employee_df.drop(drop_list, axis =1, inplace = True)
```

```python
employee_df.head(5)
```

```python
#Filtering Data for Employees Who Left

employee_left = employee_df[employee_df["Attrition"]==1]
```

```python
#Filtering Data for Employees Who are staying

employee_stayed = employee_df[employee_df["Attrition"]== 0]
```

```python
#creating correlation map

correlations = employee_df.corr(numeric_only = True)

fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(correlations, annot=True)
plt.show()
```

```python
#Selecting Categorical Features from DataFrame

X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

```python
#Applying One-Hot Encoding to Categorical Features

from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat 
```

```python
#Selecting Numerical Features from DataFrame

X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
```

```python
#Combining Categorical and Numerical Features, and Ensure Column Names are Strings
X_all = pd.concat([X_cat, X_numerical], axis=1)
X_all.columns = X_all.columns.astype(str)
print(X_all)
```

```python
#Normalizing Features Using Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
X
```

```python
y = employee_df['Attrition']
y
```

```python
#Spliting Data into Training and Testing Sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Training the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Making Predictions Using the Trained Model
y_pred = model.predict(x_test)
y_pred
```

```python
#Importing Metrics for Model Evaluation

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Calculating Model Accuracy

acs = (100*accuracy_score(y_pred, y_test))
print(f"The accuracy score is {acs:.2f}%")
```

```python
#Displaying Confusion Matrix as Heatmap

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm,annot=True)
```

```python
#classfication report
cf = classification_report(y_pred, y_test)
print(cf)
```

```python
#Training Random Forest Classifier Model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
##Calculating Model Accuracy

acs = (100*accuracy_score(y_pred, y_test))
print(f"The accuracy score is {acs:.2f}%")
```

```python
#Displaying Confusion Matrix as Heatmap

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm,annot=True)
```
